```python
data = data.drop(['Row ID', 'Order ID', 'Customer ID', 'Postal Code'], axis=1)
```

```python
data.columns
```

```python
data['Category'].unique()
```

```python
data.groupby('Category')['Profit'].sum().sort_values(ascending=False).plot(kind='bar')
plt.xticks(rotation=0)
plt.ylabel('Total Profit')
plt.title('Profit margins between different categories')
plt.show()
#Conclusion : The hypothesis is supported as technology products have the 
#highest profit margin of the three categories.
```

```python
data1 = data.groupby('Region')['Sales'].sum().sort_values(ascending=False).reset_index()
data1
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=data['Region'],y=data['Sales'], data=data1)
```

```python
reg_sales = data.groupby('Region')['Sales'].sum()
reg_sales.plot(kind='bar')
plt.xlabel("Region")
plt.ylabel("Total Sales by Region")
plt.title("Total Sales")
plt.show()
#Conclusion : The hypothesis is not supported as East region does not have the highest sales among the regions 
```

```python
data['Order Month']=pd.DatetimeIndex(data['Order Date']).month
month_sales = data.groupby('Order Month')['Sales'].sum()
month_sales.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.show()
#Conclusion : The hypothesis is supported as the Sales are higher
#during certain months of the year.
```

```python
data.columns
```

```python
total_orders_by_shipping_mode = data.groupby('Ship Mode').size()
returned_orders_by_shippinh_mode = data[data['Profit'] < 0].groupby('Ship Mode').size()
returned_per_by_shipping_mode = (returned_orders_by_shippinh_mode/total_orders_by_shipping_mode)*100
print(returned_per_by_shipping_mode)
returned_per_by_shipping_mode.plot(kind='line')
plt.title('Return Per By Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Returns Per')
plt.xticks(rotation=50)
plt.show()
#Conclusion : The hypothesis is supported as orders with same-day 
#shipping have the Lowest rate of returned products.
```

```python
data['Order Day'] = pd.DatetimeIndex(data['Order Date']).day_name()
day_sales = data.groupby('Order Day')['Profit'].sum()
day_sales.plot(kind='bar')
plt.title('Total Profit by the day of the week')
plt.xlabel('Day of the week')
plt.ylabel('Total Profit')
plt.show()
#Conclusion : The hypothesis is supported as company's profit
# is higher on weekdays compared to weekends
```
